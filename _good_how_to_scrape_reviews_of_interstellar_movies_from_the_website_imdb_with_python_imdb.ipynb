```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from tensorflow.keras.models import load_model
loaded_model = load_model("/content/drive/MyDrive/review_IMDB_sentiment_classifier1.tf")
```

```python
def predict_sentiment(text, model):
    text = [text]
    pred = model.predict(text)
    return pred
    # if pred > 0.5:
    #    return "Positive"
    # else:
    #     return "Negative"
```

```python
df=pd.read_csv("/content/drive/MyDrive/top_250_movies_imdb.csv")
```

```python
def extract_reviews(page_url):
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    reviews = []
    for review in soup.find_all('div', {'class': 'lister-item-content'}):
        rating = review.find('span', {'class': 'rating-other-user-rating'})
        if rating is not None:
            rating = int(rating.span.text)
        else:
            rating = None
        text = review.find('div', {'class': 'text'}).text.strip()
        reviews.append({'rating': rating, 'text': text})
    return reviews
```

```python
def eval(x , model):
  sum=0
  positive=0
  négative=0
  for i in range(len(x)):

    y=predict_sentiment(x[i] , model)
    if y>0.5:
      positive=positive+1
    else :
      négative=négative+1
    sc_pos=positive*100/len(x)
    sc_neg=négative*100/len(x)
  print("Le pourcentage des commentaires positives sont :" , sc_pos,"%")
  print("Le pourcentage des commentaires Négative sont :" , sc_neg,"%")
  if sc_pos>sc_neg:
    return "recommended"
  else :
    return "not recommended"

  # moy= sum/len(x)
  # if(moy<0.5):
  #  print(moy)
  #  return ("not recommended")
  # elif (moy>=0.8):
  #    print(moy)
  #    return("recommended")
  # else :
  #   print(moy)
  #   return("moyen ")
```
